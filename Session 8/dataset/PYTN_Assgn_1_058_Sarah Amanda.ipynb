# Folium Maps

```python
#!conda install -c conda-forge folium=0.5.0 --yes
!pip install folium
import folium
print('Folium installed and imported!')
```

```python
df_crime = pd.read_csv('city_of_london_street.csv')

print('Dataset downloaded and read into a pandas dataframe!')
```

1. Load dataset dari dokumen excel

```python
df_crime.head(437)
```

```python
df_crime.shape
```

```python
df_crime.Crime_type
```

```python
limit = 437
df_crime = df_crime.iloc[0:limit, :]
```

2. Menentukan latitude dan longitude values

```python
#London latitude and longitude values
latitude = 51.509865
longitude = -0.118092
```

3. Membuat titik koordinat lokasi London

```python
# create map and display it
sanfran_map = folium.Map(location=[latitude, longitude],zoom_start=12)
# display the map of London
sanfran_map
```

4. Menginput data dari excel lokasi kejahatan ke dalam peta

```python
# instantiate a feature group for the crime in the dataframe
crime = folium.map.FeatureGroup()
# loop through the 437 crimes and add each to the crime feature group
for lat, lng, in zip(df_crime.Latitude, df_crime.Longitude):
    
 crime.add_child(
 folium.CircleMarker([lat, lng],
 radius=5, # define how big you want the circle markers to be
 color='yellow',
 fill=True,
 fill_color='blue',
 fill_opacity=0.6
 )
 )
# add incidents to map
sanfran_map.add_child(crime)
```

5. Membuat penanda tampilan kategori kejahatan saat diarahkan ke atas marker

```python
# instantiate a feature group for the crime in the dataframe
crime = folium.map.FeatureGroup()
# loop through the 437 crimes and add each to the crime feature group
for lat, lng, in zip(df_crime.Latitude, df_crime.Longitude):
    
 crime.add_child(
 folium.CircleMarker([lat, lng],
 radius=5, # define how big you want the circle markers to be
 color='yellow',
 fill=True,
 fill_color='blue',
 fill_opacity=0.6
 )
 )
# add pop-up text to each marker on the map
latitudes = list(df_crime.Latitude)
longitudes = list(df_crime.Longitude)
labels = list(df_crime.Crime_type)
for lat, lng, label in zip(latitudes, longitudes, labels):
 folium.Marker([lat, lng], popup=label).add_to(sanfran_map)
# add incidents to map
sanfran_map.add_child(crime)
```

6. Membuat sederhana tampilan marker lokasi

```python
#create map and display it
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start = 12)
# loop through the 100 crimes and add each to the map
for lat, lng, label in zip(df_crime.Latitude, df_crime.Longitude,df_crime.Crime_type):
 folium.CircleMarker(
 [lat, lng],
 radius=5, # define how big you want the circle markers to be
 color='yellow',
 fill=True,
 popup=label,
 fill_color='blue',
 fill_opacity=0.6
 ).add_to(sanfran_map)
# show map
sanfran_map
```

7. Mengelompokan marker kedalam cluster yang berbeda

```python
from folium import plugins
# let's start again with a clean copy of the map of London
sanfran_map = folium.Map(location = [latitude, longitude], zoom_start = 12)
# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(sanfran_map)
# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(df_crime.Latitude, df_crime.Longitude,df_crime.Crime_type):
 folium.Marker(
 location=[lat, lng],
 icon=None,
 popup=label,
 ).add_to(incidents)
# display map
sanfran_map
```
